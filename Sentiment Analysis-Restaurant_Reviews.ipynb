# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
wc=WordNetLemmatizer()
```

```python
corpus=[]
```

```python
#preprocessing all the reviews
for i in range(0,1000):
    review=dataset['Review'][i]
    review=re.sub('[^a-zA-Z]'," ",review)
    review=review.lower()
    #tokenization
    review=review.split()
    #removing stop words and applying stemming
    d=[]
    for word in review:
        if(word not in new_sw):
            d.append(wc.lemmatize(word))
    review=" ".join(d)
#all the preprocessed review is saved in corpus
    corpus.append(review)
```

# Splitting the data

```python
#y = Liked..Liked is a target
y=dataset['Liked']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_tfidf,y,test_size=0.2,random_state=42)
```

```python
#no need of scaling as the data is in 0 and 1 range mostly
```

#  Modelling

# Building the model using NN

```python
from keras.models import Sequential #to initialize
from keras.layers import Dense#to add layers
```

# Init the NN model

```python
model=Sequential()
```

# Add Input

```python
x_train.shape[1] #800 reviews
```

```python
model.add(Dense(input_dim=x_train.shape[1],kernel_initializer='random_uniform',activation='relu',units=20))
```

# Add Hidden layer

```python
model.add(Dense(kernel_initializer='random_uniform',activation='relu',units=20))
```

# Add Output layer

```python
#to overcome the overfitting problems, we do regularization(l1//l2)
#regualrization will apply on the last layer
```

```python
from keras import regularizers
```

```python
l1=regularizers.l2(l=0.1)
```

```python
#the output is binary..so we go for sigmoid
#unit=1..output only one(0 or 1)
model.add(Dense(kernel_initializer='random_uniform',activation='sigmoid',units=1,kernel_regularizer=l1))
```

# Compile the model

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

# Train the model

```python
#not a big range data..so gng for 10 epochs
#count of each batch..every 32 rows of data...we're updating our weights
model.fit(x_train,y_train,epochs=10,batch_size=32)
```

# prediction

```python
y_pred=model.predict(x_test)
```

```python
np.set_printoptions(precision=2,suppress=True)
```

```python
y_pred
#now it's in the form of probability
```

```python
#we need to convert into classes before predicting
y_pred=model.predict_classes(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
print("Test Accuracy score:",accuracy_score(y_test,y_pred)*100,"%")
```

```python
#training accuracy high...testing low..(over fitting problem)

#later we gor for regularization to overcome the over fitting problem
```
